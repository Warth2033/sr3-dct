```python
import matplotlib.pyplot as plt
import networkx as nx

def draw_network():
    graph = nx.DiGraph()

    nodes = [
        "Input",
        "Downsampling Block 1",
        "Downsampling Block 2",
        "Downsampling Block 3",
        "Middle Block",
        "Upsampling Block 1",
        "Upsampling Block 2",
        "Upsampling Block 3",
        "Output"
    ]

    for node in nodes:
        graph.add_node(node)

    edges = [
        ("Input", "Downsampling Block 1"),
        ("Downsampling Block 1", "Downsampling Block 2"),
        ("Downsampling Block 2", "Downsampling Block 3"),
        ("Downsampling Block 3", "Middle Block"),
        ("Middle Block", "Upsampling Block 1"),
        ("Upsampling Block 1", "Upsampling Block 2"),
        ("Upsampling Block 2", "Upsampling Block 3"),
        ("Upsampling Block 3", "Output")
    ]

    for edge in edges:
        graph.add_edge(*edge)

    pos = nx.spring_layout(graph)
    nx.draw(graph, pos, with_labels=True, node_size=4000, node_color="lightblue", font_size=8, font_color="black", font_weight="bold")
    plt.title("Network Structure")
    plt.show()

# Run the function to draw the network
draw_network()
```
